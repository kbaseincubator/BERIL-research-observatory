# NB01: Gene Family Extraction from kbase_ke_pangenome

**Requires BERDL JupyterHub** — this notebook queries 132.5M gene clusters via Spark SQL.

Outputs: `data/species_gene_families.csv`, `data/species_taxonomy.csv`, `data/phenazine_operon_species.csv`

Run equivalent script: `python src/01_extract_gene_families.py`

## Gene Family Definitions

23 gene families across 4 functional groups:
- **P-acquisition (9):** phoA, phoD, pstA/B/C/S, phnC/D/E
- **N-fixation (3):** nifH, nifD, nifH(Pfam)
- **Metal-handling (4):** copA, corA, feoB, HMA
- **Phenazine biosynthesis (7):** phzA/B/D/F/G/S/M

```python
import pandas as pd
import os
DATA_DIR = os.path.join('..', 'data')
```

```python
# Load pre-extracted results (Spark query already executed)
df = pd.read_csv(os.path.join(DATA_DIR, 'species_gene_families.csv'))
print(f'Total species: {len(df)}')
print(f'Columns: {len(df.columns)}')
df.head()
```

## Summary Statistics

```python
print(f"P-acquisition: {df['has_P_acquisition'].sum()} ({df['has_P_acquisition'].mean()*100:.1f}%)")
print(f"N-fixation: {df['has_N_fixation'].sum()} ({df['has_N_fixation'].mean()*100:.1f}%)")
print(f"Metal-handling: {df['has_metal_handling'].sum()} ({df['has_metal_handling'].mean()*100:.1f}%)")
print(f"Phenazine operon (>=3 phz genes): {df['has_phz_operon'].sum()} ({df['has_phz_operon'].mean()*100:.2f}%)")
print(f"PhzF only (broad family): {df['has_phzF'].sum()}")
print()
print(f"P + Metal: {((df['has_P_acquisition']==1) & (df['has_metal_handling']==1)).sum()}")
print(f"N + Metal: {((df['has_N_fixation']==1) & (df['has_metal_handling']==1)).sum()}")
print(f"Phz operon + Metal: {((df['has_phz_operon']==1) & (df['has_metal_handling']==1)).sum()}")
print(f"All four: {((df['has_P_acquisition']==1) & (df['has_N_fixation']==1) & (df['has_metal_handling']==1) & (df['has_phz_operon']==1)).sum()}")
```
