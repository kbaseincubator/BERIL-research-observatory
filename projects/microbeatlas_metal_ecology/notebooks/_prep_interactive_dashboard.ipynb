```python
spark = get_spark_session()
```

```python
import pandas as pd
from pathlib import Path
```

```python
DATA = Path("/home/hmacgregor/BERIL-research-observatory/projects/microbeatlas_metal_ecology/data")
```

```python
df = spark.sql("""
    SELECT sample_id,
           LatitudeParsed,
           LongitudeParsed,
           Env_Level_1
    FROM arkinlab_microbeatlas.sample_metadata
""").toPandas()
df.to_csv(DATA / 'sample_latlon_env.csv', index=False)
```

```python
# Sample-level OTU richness
sample_richness = spark.sql("""
    SELECT
        sample_id,
        COUNT(DISTINCT otu_id) AS otu_richness
    FROM arkinlab_microbeatlas.otu_counts_long
    GROUP BY sample_id
""").toPandas()
sample_richness.to_csv(DATA / 'sample_otu_richness.csv', index=False)
```
